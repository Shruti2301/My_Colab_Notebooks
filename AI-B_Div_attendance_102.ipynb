```
# Import libraries
import pandas as pd
import glob
import numpy as np
```

```
path = r'/content/drive/MyDrive/Classroom/attendance' # path to sheets
all_files = glob.glob(path + "/*.csv") #store path of each sheet

df = []
li=[]

for filename in all_files:
    df = pd.read_csv(filename,encoding = 'utf-16',sep='\t') #read sheets one by one
    df = df.drop_duplicates(['Full Name']) #drop duplicate names from each sheet
    li.append(df) #store the new list and keep appending
df = pd.concat(li, axis=0, ignore_index=True) #concat the dataframe properly
total = pd.DataFrame(df.pivot_table(index = ['Full Name'], aggfunc ='size')) #count the number of times a name occurs
total.sort_values(by=['Full Name'], inplace=True) #sort the names in ascending order
print(total)
```

```
#Export CSV
total.to_csv('/content/drive/MyDrive/Classroom/attendance/TotalAttendanceBdiv.csv')
```
