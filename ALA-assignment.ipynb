```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10,10,1)
y = np.arange(-20,20,2)

f = 3*(x**2) + 2*x*y
L = 8*x + 2*y - 5 #Equation approximated at point (1,1)

fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot(x,y,f)
ax.plot(x,y,L)
plt.show()
```
