```python
import numpy as np
import matplotlib.pyplot as plt

# Function to generate synthetic data with normal distribution
def generate_data(mean, std, size):
    return np.random.normal(mean, std, size)

# Function to simulate dataset shift types
def simulate_dataset_shift(dataset_type):
    if dataset_type == 'covariate':
        return generate_data
    elif dataset_type == 'label':
        return lambda mean, std, size: np.random.choice([-1, 1], size=size)
    elif dataset_type == 'both':
        return lambda mean, std, size: np.random.normal(mean, std, size) * np.random.choice([-1, 1], size=size)
    else:
        raise ValueError('Invalid dataset shift type.')

# Function to plot the data distribution
def plot_data(data, title):
    plt.hist(data, bins=50)
    plt.title(title)
    plt.show()

# Generate synthetic data with different distribution shifts
original_data = generate_data(0, 1, 1000)
covariate_shift_data = simulate_dataset_shift('covariate')(0, 1, 1000)
label_shift_data = simulate_dataset_shift('label')(0, 1, 1000)
both_shift_data = simulate_dataset_shift('both')(0, 1, 1000)

# Plot the data distributions
plot_data(original_data, 'Original data')
plot_data(covariate_shift_data, 'Covariate shift data')
plot_data(label_shift_data, 'Label shift data')
plot_data(both_shift_data, 'Both shift data')
```
